```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print(" ==: zomzto sales analysis :==")
print("\n sales data shape",np.shape(sales_data))
print("\n first 3 resturent data",sales_data[:3])
```

```python
# Total sales per year
print("total sales per year := \n",np.sum(sales_data,axis=0))
yearly_total=np.sum(sales_data[:,1:],axis=0)
print("year_total  :==\n",yearly_total)
total = np.sum(yearly_total)
print("total sales :==\n",total)

# MINIMUM sales per resturent
min_sales = np.min(sales_data[:,1:],axis=1)
print("minimum sales per resturent :==\n",min_sales)


#MAXIMUM sales per year
max_year_sales =np.max(sales_data[:,1:],axis=0)
print("maximum sales per year :==\n",max_year_sales)

#Average sales per restaurent
avg_sales = np.mean(sales_data[:,1:],axis=1)
print("average sales per restaurent :==\n",avg_sales)

#Average sales per year
avg_year = np.mean(sales_data[:,1:],axis=0)
print("average sales per year ;=\n",avg_year)


"""
 note is that in axis=0 = column and axis =1= rows-- and by default is 0
 but in function[x,y] then x= rows and y=clmns
 """


cumulative_sales = np.cumsum(sales_data[:,1:],axis=0)
print("cumulative sales :==\n",cumulative_sales)

plt.figure(figsize=(10, 6))
plt.plot(np.mean(cumulative_sales,axis=0))
plt.title("Cumulative Sales Over Years")
plt.xlabel("Year")
plt.ylabel("Cumulative Sales")
plt.grid(True)
plt.show()
```

```python
vecto1 = np.array([1,2,3,4,5])
vecto2 = np.array([6,7,8,9,2])
print("Adition of vector : ",vecto1+vecto2)
print("Adition of multiplication : ",vecto1*vecto2)
print("Adition of dot : ",np.dot(vecto1,vecto2))
angle = np.arccos(np.dot(vecto1,vecto2)/(np.linalg.norm(vecto1)*np.linalg.norm(vecto2)))
print("Angle between vectors : ",np.degrees(angle))
```
