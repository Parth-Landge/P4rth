**bivariate:- numerical - categorical**

```python
hr_sal = df[df["dept"] =="HR"]["salary"]
hr_sal
```

```python
it_sal =df[df["dept"]=="IT"]["salary"]
it_sal
```

```python
fin_sal =df[df["dept"]=="Finance"]["salary"]
fin_sal
```

```python
plt.boxplot([hr_sal,it_sal,fin_sal],labels=["HR","IT","Finance"])
plt.grid()
plt.show()
```

```python
#pie chart
```

```python
salary_by_dept = df.groupby("dept")["salary"].sum()
salary_by_dept
```

```python
mark_sal=df[df["dept"]=="Marketing"]["salary"]
mark_sal
```

```python
plt.pie(salary_by_dept,labels=salary_by_dept.index,autopct=  "%1.2f")
plt.show()
```

```python
hr_mean = sum(hr_sal)/len(hr_sal)
it_mean = sum(it_sal)/len(it_sal)
fin_mean = sum(fin_sal)/len(fin_sal)
mark_mean = sum(fin_sal)/len(fin_sal)
```

```python
plt.bar(["HR","IT","Finance","Marketing"],[hr_mean,it_mean, fin_mean,mark_mean],color =["gray","pink","cyan","purple"])
plt.grid()
plt.show()
```

```python
df["experience"] = [1,2,4,1.5,5,8,12,10,4,1,1.2,2,3,1.8,5.5,4.5,3.2,7,8,10]
df.head()
```

```python
plt.scatter(df["Age"],df["salary"],s =df["experience"]*30,color = "black",edgecolors="green")
plt.title("age vs salary vs experince")
plt.xlabel("Age")
plt.ylabel("salary")
plt.show()
```

```python
# 2 numerical and 1 categorical
```

```python
plt.scatter(df["Age"],df["salary"], c = df["dept"].map({"HR":"yellow","IT":"blue","Finance":"orange","Marketing":"black"}),label =["HR","IT","Finance","Marketing"])
plt.xlabel("Age")
plt.ylabel("IT")
plt.legend()
plt.show()
```

**object oriented API**

```python
fig,axs = plt.subplots(1,3,figsize=(15,5))
#line plot 
axs[0].plot(sort_age["Age"],df["salary"],color =  "red",linestyle=":")
axs[0].set_xlabel("Age")
axs[0].set_ylabel("salary")
axs[0].set_title("1st")
#histogram
axs[1].hist(df["salary"],bins = 5,color ="cyan")
axs[1].set_title("histo")
axs[1].set_xlabel("salary")
axs[1].set_ylabel("frquemcy")

#boxplot
axs[2].boxplot(df["salary"])
axs[2].set_title("boxplpot")
axs[2].set_xlabel ("salary")
plt.show()
```
